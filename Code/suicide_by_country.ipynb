```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
suicide_file = "../Raw Data/suicide_rates.csv"
suicide_df = pd.read_csv(suicide_file)
suicide_df.head()
```

```python
country_file = "../Raw Data/Country_Data.csv"
country_df = pd.read_csv(country_file)
country_df.head()
country_df = country_df.rename(columns={"Country Name":"country","Country Code":"country_code"})
country_df.head()
```

```python
suicide_cols = ['LOCATION','TIME','Value']
suicide_transformed = suicide_df[suicide_cols].copy()

suicide_transformed = suicide_transformed.rename(columns={"LOCATION":"country_code","TIME":"year","Value":"suicide_rate"})

```

```python
suicide_transformed.head()
```

```python
suicide_by_country = pd.merge(suicide_transformed, country_df, on= "country_code")
```

```python
suicide_by_country.head()
```

```python
suicide_by_country_clean = suicide_by_country[['country','country_code','year','suicide_rate']]
```

```python
suicide_by_country_clean.head()
```

```python
output_data_file = "../Resources/suicide_clean_df.csv"
suicide_by_country_clean.to_csv(output_data_file, index = False)
```
