```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../Raw Data/Human Development Index (HDI).csv"

# Read Purchasing File and store into Pandas data frame
HDI_data = pd.read_csv(file_to_load, encoding= 'unicode_escape')

#display dataframe
HDI_data.head()
```

```python
#delete top 4 rows 
HDI = HDI_data.iloc[4:]
HDI.reset_index(inplace=True)
HDI.head()
```

```python
HDI.drop(['index'], axis=1)
```

```python
HDI.columns = HDI.iloc[0]
HDI.head()
```

```python
HDI = HDI.iloc[1:]
HDI.head()
```

```python
row= HDI.tail(15)
print(row)
```

```python
HDI = HDI.iloc[:-15]
row= HDI.tail(15)
print(row)
```

```python
HDI=HDI.dropna(axis=1, how="all")
HDI.head()
```

```python
years= list(HDI.columns)[4:33]
years
```

```python
HDI=HDI.melt(id_vars=['HDI Rank','Country'], value_vars=years)
HDI
```
