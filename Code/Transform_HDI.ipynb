## Import Dependencies

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

## Store CSV in Datafram

```python
# File to Load
file_to_load = "../Raw Data/Human Development Index (HDI).csv"

# Read HDI File and store into Pandas data frame
HDI_data = pd.read_csv(file_to_load, encoding= 'unicode_escape')

#display dataframe
HDI_data.head()
```

## Delete unnecessary row and columns

```python
#delete top 4 rows using loc
HDI = HDI_data.iloc[4:]
HDI.reset_index(inplace=True)
HDI.head()
```

```python
#dropping index
HDI.drop(['index'], axis=1)
```

```python
#making first rwo the header
HDI.columns = HDI.iloc[0]
HDI.head()
```

```python
#removing first row after made header
HDI = HDI.iloc[1:]
HDI.head()
```

```python
#checking if any unnecessary rows at the end of the file 
row= HDI.tail(20)
print(row)
```

```python
#removing extra rows from the bottom
HDI = HDI.iloc[:-17]
row= HDI.tail(15)
print(row)
```

```python
#removing all the columns with all values as NaN
HDI=HDI.dropna(axis=1, how="all")
HDI.head()
```

## Melt Years to add Year as Row 

```python
#get the list of years in the dataframe
years= list(HDI.columns)[4:33]
years
```

```python
HDI=HDI.melt(id_vars=['HDI Rank','Country'], value_vars=years)
HDI
```

## Rename Columns

```python
HDI=HDI.rename(columns={0: "year","HDI Rank": "hdi_rank","Country": "country","value": "value"})
HDI.head()
```

# Eliminate first empty character of Country

```python
#what is the first character in the country
HDI.country[1]
```

```python
HDI['country'] = HDI['country'].str[1:]
HDI.head()
```

# Order by country then year

```python
HDI = HDI.sort_values(by=['country', 'year'])
HDI
```

# Last NaN check

```python
#checking if there are any NaN values missed out
is_NaN = HDI.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = HDI[row_has_NaN]

print(rows_with_NaN)
```

```python
HDI = HDI[HDI['value'] != ".."]
print('\nCleanes dataframe :\n', HDI)
```

# Bring Country Data in line with other data
The GDP and suicide rate data sources are the same and hence have the same naming conventions, the HDI data has a different data source therefore the country columns needs to be brought in line with the other two data sources since the country is one of the Primary Keys in the data base

```python
#load the data set which compares the country names across two data sources
link_country = "../Raw Data/HDI_countries_CD.csv"
country_data = pd.read_csv(link_country)
country_data.head()
```

```python
#drop countries that don't have any comparision to offer (like American Samoa)
country_data = country_data.dropna()
country_data.head()
```

```python
#identify/shortlist countries that have different names in the GDP/Suicide Rate Data Set
country_data = country_data[country_data["country(HDI)"] != country_data["Country Name(GDP/SR)"]]
country_data
```

```python
#replace the values
for z in range(len(HDI)):
    for x in range(len(country_data)):
        if str(HDI.iloc[z,1])==str(country_data.iloc[x,0]):
             HDI.iloc[z,1]=country_data.iloc[x,1]
```

```python
#check if replacement worked 
rslt_df = HDI[HDI['country'] == "Bahamas, The"]
  
print('\nResult dataframe :\n', rslt_df)
```

```python
HDI=HDI.reset_index(drop=True)
HDI.head()
```
