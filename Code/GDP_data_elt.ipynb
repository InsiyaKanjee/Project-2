```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
link_gdp = "../Raw Data/GDP_capita.csv"
link_country = "../Raw Data/Country_Data.csv"
# Read GDP  File and store into Pandas data frame
gdp_data = pd.read_csv(link_gdp)
gdp_data.head()
```

```python
# Read country data file and store into Pandas data frame
country_data = pd.read_csv(link_country)

country_data.head()
```

# Cleaning the GDP data

```python
#Selecting and renaming relevant columns in GDP df
gdp_clean_data = gdp_data[['LOCATION','TIME','Value']]
gdp_clean_data = gdp_clean_data.rename(columns = {"LOCATION": "country_code", "TIME": "year","Value": "gdp_per_capita"})
gdp_clean_data.head
```

```python
#Selecting and renaming relevant columns in country df

country_clean_data = country_data.rename(columns = {"Country Code": "country_code", "Country Name": "country"})
country_clean_data.head
```

```python
#adding full country name
gdp_df = pd.merge(gdp_clean_data,country_clean_data, on = "country_code", how = "inner")
gdp_df
```

```python
#droppiging NAN values
gdp_clean_df = gdp_df.dropna(how = "any")
gdp_clean_df = gdp_clean_df[["country","year","gdp_per_capita"]]
gdp_clean_df
```

```python
#save data as csv
output_data_file = "../Resources/gdp_clean_df.csv"
gdp_clean_df.to_csv(output_data_file, index = False)
```
